```python
#Fill in missing values
df["total_bedrooms"].fillna(0, inplace=True)
```

**Make median_house_value binary**

```python
#Above Average
df["above_average"] = df["median_house_value"]>df["median_house_value"].mean()
df["above_average"] = df["above_average"].astype(int)    
```

```python
from sklearn.model_selection import train_test_split
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=42)
```

```python
#Reset indices
df_train_full.reset_index(drop=True)
df_train.reset_index(drop=True)
df_val.reset_index(drop=True)
df_test.reset_index(drop=True)
```

```python
#Obtain y for logistic regression
y_train_full = df_train_full["above_average"].values
y_train = df_train["above_average"].values
y_val = df_val["above_average"].values
y_test = df_test["above_average"].values

#Obtain new y for linear regression (Q6)
y_train_linear = df_train["median_house_value"].values
y_val_linear = df_val["median_house_value"].values
```

```python
#Drop y from training and test sets
df_train_full.drop(["median_house_value", "above_average"], axis=1, inplace=True)
df_train.drop(["median_house_value", "above_average"], axis=1, inplace=True)
df_val.drop(["median_house_value", "above_average"], axis=1, inplace=True)
df_test.drop(["median_house_value", "above_average"], axis=1, inplace=True)
```

**Question 3**

```python
#Mutual info score for Ocean Proximity
from sklearn.metrics import mutual_info_score
round(mutual_info_score(df_train["ocean_proximity"],y_train),2)
```

**Question 4**

```python
#One hot encoding
from sklearn.feature_extraction import DictVectorizer
train_dict = df_train.to_dict(orient='records')
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
X_train = dv.transform(train_dict)
X_train.shape
```

```python
#Train logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver="liblinear", C=1.0, max_iter=1000, random_state=42)
model.fit(X_train, y_train)
```

```python
#Accuracy on validation set

#One hot encoding to obatain X_val
val_dict = df_val.to_dict(orient='records')
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
X_val = dv.transform(val_dict)

#Calculate predicted y
y_val_pred = model.predict_proba(X_val)[:,1]
y_val_pred

#Accuracy
y_val_above_average = y_val_pred>0.5
accuracy = (y_val == y_val_above_average).mean()
round(accuracy, 2)
```

**Question 5**

```python
for var in df_train.columns:
    
    df_train_adjusted = df_train.drop(var, axis=1)
    train_adjusted_dict = df_train_adjusted.to_dict(orient='records')
    dv = DictVectorizer(sparse=False)
    dv.fit(train_adjusted_dict)
    X_train_adjusted = dv.transform(train_adjusted_dict)
    
    #Train logistic regression
    model_adjusted = LogisticRegression(solver="liblinear", C=1.0, max_iter=1000, random_state=42)
    model_adjusted.fit(X_train_adjusted, y_train)
    
    #One hot encoding to obatain X_val
    df_val_adjusted = df_val.drop(var, axis=1)
    val_adjusted_dict = df_val_adjusted.to_dict(orient='records')
    dv = DictVectorizer(sparse=False)
    dv.fit(val_adjusted_dict)
    X_val_adjusted = dv.transform(val_adjusted_dict)
        
    #Calculate predicted y
    y_val_pred_adjusted = model_adjusted.predict_proba(X_val_adjusted)[:,1]
    y_val_pred_adjusted

    #Accuracy
    y_val_above_average_adjusted = y_val_pred_adjusted>0.5
    accuracy_adjusted = (y_val == y_val_above_average_adjusted).mean()
    
    print("Feature eliminated:", var, "Difference in accuracy:", accuracy-accuracy_adjusted)
```

**Question 6**

```python
#define function to calculate RMSE
def rmse(y, y_pred):
    error = y_pred - y
    mse = (error ** 2).mean()
    return np.sqrt(mse)
```
